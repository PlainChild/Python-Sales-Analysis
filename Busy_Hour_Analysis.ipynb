```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/PlainChild/Python-Sales-Analysis/refs/heads/main/Order_details(masked).csv')
df.head(5)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df[df.duplicated()]
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], dayfirst=True)
df.dtypes
```

```python
df['Hour'] = (df['Transaction Date']).dt.hour
df.head(5)
```

```python
busyhour = df['Hour'].value_counts()
busyhour
```

```python
busyhour1 = busyhour.sort_index()
busyhour2 = []

for i in range(0,23):
    busyhour2.append(i)


plt.figure(figsize=(12, 6))

plt.title('Sales Per Hour',
          fontdict={'fontname': 'monospace', 'fontsize': 30}, y=1.05)

plt.ylabel("Number Of Purchases Made", fontsize=18, labelpad=20)
plt.xlabel("Hour", fontsize=18, labelpad=20)
plt.plot(busyhour2, busyhour1, color='m')
plt.grid()
plt.show()

```
